```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("homework_3.1.csv")

for col in ["value1", "value2", "value3"]:

    temp = df.copy()

    temp["post"] = (temp["time"] >= 50).astype(int)

    X = sm.add_constant(
        temp[["time", "post"]]
    )

    model = sm.OLS(temp[col], X).fit()

    print("\n", col)
    print(model.params)
```

```python
for col in ["value1", "value2", "value3"]:

    temp = df.copy()

    temp["post"] = (temp["time"] >= 50).astype(int)

    temp["time_post"] = (
        temp["time"] * temp["post"]
    )

    X = sm.add_constant(
        temp[["time", "post", "time_post"]]
    )

    model = sm.OLS(temp[col], X).fit()

    print("\n", col)
    print(model.params)
```

```python
for group in ["Group1", "Group2"]:

    temp = df.copy()

    temp["post"] = (
        temp["time"] >= 50
    ).astype(int)

    temp["interaction"] = (
        temp["treated"]
        * temp["post"]
    )

    X = sm.add_constant(
        temp[["treated",
              "post",
              "interaction"]]
    )

    model = sm.OLS(
        temp[group],
        X
    ).fit()

    print(group)
    print(model.params)
```
